Existe missing na coluna *Data de abertura*

```python
lic_dados['Data Abertura'].isnull().value_counts()
```

```python
lic_dados.columns
```

```python
lic_dados['Data Abertura'] = pd.to_datetime(lic_dados['Data Abertura'], errors='coerce', dayfirst=True)
```

```python
lic_dados['Data Resultado Compra'] = pd.to_datetime(lic_dados['Data Resultado Compra'], errors='coerce')
#Convertendo Data Resultado Compra para datetime
```

```python
lic_dados_filtrados = lic_dados[
    lic_dados['Data Abertura'].isnull() & (lic_dados['Data Resultado Compra'].dt.year < 2018)
]
#Criando o filtro para encontrar data de abertura faltando e data resultado compra antes de 2018
```

```python
print(lic_dados_filtrados)
#Não foram encontradas linhas no DataFrame lic_dados que atendam aos critérios de pesquisa
 #(Data Abertura faltando e Data Resultado Compra menor que 2018).
```

Como não tinham as datas faltando e data resultado compra antes de 2018, por esse motivo, optei por colocar a data padrão de substituição como 01.01.2018

```python
lic_dados['Data Abertura'] = lic_dados['Data Abertura'].fillna(pd.to_datetime('01/01/2018', dayfirst=True))
```
